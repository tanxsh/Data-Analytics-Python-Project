```python
import pandas as pd

years = [2020, 2021, 2022, 2023, 2024, 2025]

all_crashes = []
all_parties = []

# Load files
for year in years:
    # skip to handle bad lines
    c_df = pd.read_csv(f"{year}_Crashes.csv", low_memory=False, on_bad_lines='skip')
    p_df = pd.read_csv(f"{year}_Parties.csv", low_memory=False, on_bad_lines='skip')
    
    # Year df
    c_df['Year'] = year
    p_df['Year'] = year
    
    all_crashes.append(c_df)
    all_parties.append(p_df)

# Combine all years
df_crashes_final = pd.concat(all_crashes, ignore_index=True)
df_parties_final = pd.concat(all_parties, ignore_index=True)

# Save 
df_crashes_final.to_csv("merged_c_merge.csv", index=False)
df_parties_final.to_csv("merged_p_merge.csv", index=False)
```
